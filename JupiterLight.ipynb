```python
from PIL import ImageGrab
import tkinter as tk
import time
import asyncio



root = tk.Tk()
Screen_Res = (root.winfo_screenwidth(), root.winfo_screenheight())

class light():
    def __init__(self, x, y, Xr, Yr):
        self.x = x
        self.y = y
        self.Xr = Xr
        self.Yr = Yr
        self.rgb = ()

    async def average2(self):
        sqr = ImageGrab.grab(bbox = (self.x, self.y, self.x + self.Xr, self.y + self.Yr))
        px = sqr.load()
        totalRGB = (0, 0, 0)
        for x in range(self.Xr):
            for y in range(self.Yr):
                totalRGB = tuple(map(sum, zip(px[x,y], totalRGB))) 
        RGB = tuple(ti/(self.Xr * self.Yr) for ti in totalRGB)
        self.rgb = RGB

    #def most common (so much more complex)

    def __str__(self):
        return str(self.rgb)

#initializing light objects
lights = []

def GCF(x, y):
    while(y):
        x, y = y, x % y
    return abs(x)


def init():
    var = GCF(Screen_Res[0], Screen_Res[1])

    for x in range(int(Screen_Res[0]/var)):
        l = light(x*var, 0, var, var)
        lights.append(l)

    for y in range(int((Screen_Res[1] - var)/var)):    
        l = light(Screen_Res[0] - var, (y +1)* var, var, var)
        lights.append(l)

    for X in range(int((Screen_Res[0] - var)/var), 0, -1):
        l = light((X - 1 )*var, Screen_Res[1] - var, var, var)
        lights.append(l)

    for Y in range(int((Screen_Res[1] - var)/var), 1, -1):
        l = light(0, (Y - 1) * var, var, var)
        lights.append(l)

    for i in lights:
        print(i.x, i.y)

#I got all my lights!

init()
start2 = time.time()

#lets do this async
async def main():
    #async with asyncio.TaskGroup() as tg:
    tsks = []
    for i in lights:
        tsks.append(asyncio.create_task(i.average2()))
        
    [await t for t in tsks]


await main()
#asyncio.run(main())
print(time.time() - start2)

start3 = time.time()
#doing it normally
for i in lights:
    i.average2()

print(time.time() - start3)
```
